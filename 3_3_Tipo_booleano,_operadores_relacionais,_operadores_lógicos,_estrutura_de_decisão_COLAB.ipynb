```python
#def calculadoraImc():
print("----- Calculadora Mágica: Salva sua prova e diverte a mulecada -----")
print("Calcule o seu IMC")
massa = float(input("Digite o seu peso abaixo:\n"))
altura = float(input("Digite sua altura em metros abaixo:\n"))
imc = massa / altura**2
if imc <= 18.5 :
  print("O valor do seu IMC é igual a: ","%.2f" %imc,". Você está abaixo do peso", sep = "")
if imc >= 18.6 and imc <= 24.9:
  print("O valor do seu IMC é igual a: ","%.2f" %imc,". Você está no peso ideal", sep = "")
if imc >= 25 and imc <= 29.9:
  print("O valor do seu IMC é igual a: ","%.2f" %imc,". Você está no sobrepeso", sep = "")
if imc >= 30 and imc <= 34.9:
  print("O valor do seu IMC é igual a: ","%.2f" %imc,". Você está na obesidade", sep = "")
if imc >= 35 and imc <= 39.9:
  print("O valor do seu IMC é igual a: ","%.2f" %imc,". Você está na obesidade severa", sep = "")
if imc > 39.9:
  print("O valor do seu IMC é igual a: ","%.2f" %imc,". Você está na obesidade mórbida", sep = "")
```
