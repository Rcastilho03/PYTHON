```python
num1 = 30
num2 = 14
#variáveis para todas as operações aritméticas e suas atribuições
plus = num1 + num2
less = num1 - num2
times = num1 * num2
division = num1 / num2
wholeDivision = num1 // num2
moduleDivision = num1 % num2
square = num1 ** num2

#print do resultado para cada variável criada
print("The addition of" ,num1,"with",num2,"is equal to",plus)
print("The substraction of" ,num1,"with",num2,"is equal to",less)
print("The multiplication of" ,num1,"with",num2,"is equal to",times)
print("The division of" ,num1,"with",num2,"is equal to",division)
print("The whole division of" ,num1,"with",num2,"is equal to",wholeDivision)
print("The division rest of" ,num1,"with",num2,"is equal to",moduleDivision)
print(num1,"square to",num2,"is equal to",square)

```

```python
def calculadoraMruv():
  print("----- Calculadora Mágica: Salva sua prova e diverte a mulecada -----")
  print("Calcule o MRUV - Movimento Retilíneo Uniforme Variado")
  s0 = input("Digite a distância inicial (s0) abaixo:\n")
  v0 = input("Digite a velocidade inicial (v0) abaixo:\n")
  t = input("Digite a variação do tempo (t) abaixo:\n")
  a = input("Digite a aceleração média (a) abaixo:")
  s = float(s0) + float(v0) * float(t) + ((float(a) * float(t)**2)/2)
  print("A distância percorrida foi de ",s,"m",sep = "")
```

```python
def calculadoraHorario():
  print("----- Calculadora Mágica: Salva sua prova e diverte a mulecada -----")
  print("Calcule o Horário")
  region = int(input('''
  Onde você mora? (Where do you live?)
  Se for no Brasil, digite 1 (If it's in Brazil, type 1)
  Se for nos Estados Unidos, digite 2 (If it's in Unitade States, type 2)
  '''))
  if region == 1:
    horarioBase = int(input("Digite quantos segundos são:\n"))
  if region == 2:
    horarioBase = int(input("Type how many seconds are:\n"))


  horas = horarioBase//3600;  horarioSemHora = horarioBase % 3600
  minutos = horarioSemHora//60; segundos = horarioSemHora % 60

  if horas > 12 and region == 2 :
    horas -= 12
    print("\nNow is ",str(horas),":",str(minutos),":",str(segundos)," PM",sep = "")
  elif horas <= 12 and region == 2:
    print("\nNow is ",str(horas),":",str(minutos),":",str(segundos)," AM",sep = "")
  else:
    print("\nAgora são ",str(horas),":",str(minutos),":",str(segundos),sep = "")
```
